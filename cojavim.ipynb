```python
import random

def main():
    print("🧟‍♂️ Vítej v zombie hře!")
    print("Jsi v opuštěném městě a proti tobě stojí zombík.")
    print("Přežij, nebo se přidej k nim...\n")

    hrac_hp = 30
    zombie_hp = 25
    kolo = 1

    while hrac_hp > 0 and zombie_hp > 0:
        print(f"\n===== KOLO {kolo} =====")
        print(f"Tvoje HP: {hrac_hp} | HP zombíka: {zombie_hp}")
        print("1) Zaútočit")
        print("2) Léčit se")
        print("3) Zkusit utéct")

        volba = input("Vyber akci (1/2/3): ")

        if volba == "1":
            # útok hráče
            dmg = random.randint(4, 10)
            print(f"\n💥 Útočíš a dáváš zombíkovi {dmg} dmg!")
            zombie_hp -= dmg

        elif volba == "2":
            # léčení
            heal = random.randint(5, 12)
            hrac_hp += heal
            print(f"\n💊 Léčíš se o {heal} HP. Aktuální HP: {hrac_hp}")

        elif volba == "3":
            # pokus o útěk
            sance = random.randint(1, 100)
            if sance <= 40:
                print("\n🏃‍♂️ Podařilo se ti utéct! Jsi naživu... zatím.")
                return
            else:
                print("\n❌ Nepodařilo se ti utéct!")

        else:
            print("\nNeplatná volba, ztrácíš čas!")

        # pokud je zombík mrtvý, konec
        if zombie_hp <= 0:
            print("\n🎉 Zabil jsi zombíka! Přežil jsi… aspoň pro dnešek.")
            break

        # útok zombíka
        dmg_z = random.randint(3, 9)
        hrac_hp -= dmg_z
        print(f"🧟‍♂️ Zombík tě kousl za {dmg_z} dmg! Tvoje HP: {hrac_hp}")

        if hrac_hp <= 0:
            print("\n☠️ Zombík tě dostal. Hra končí.")
            break

        kolo += 1

if __name__ == "__main__":
    main()
```
